Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Scatter Plot is a plot or graph which uses co-ordinates to display values for typically two variables for a set of data.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X side of the Graph")
plt.ylabel("Y side of the Graph")
plt.title("Title of the Graph")
plt.grid()
```

Sub plot() function is used to plot multiple plots or sub plots in a single figure.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(10,10))
# Line plot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue', marker='o')
plt.title("Line 1")
plt.xlabel("X values")
plt.ylabel("Y values")
# Line plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y1, color='blue', marker='o')
plt.title("Line 2")
plt.xlabel("X values")
plt.ylabel("Y values")
# Line plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='blue', marker='o')
plt.title("Line 3")
plt.xlabel("X values")
plt.ylabel("Y values")
# Line plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='blue', marker='o')
plt.title("Line 4")
plt.xlabel("X values")
plt.ylabel("Y values")
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Bar plot is used to represent a numerical value with respect with categorical value in a graph.The height of the bar represents its numerical value.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company and Profit")
```

```python
plt.barh(company, profit)
plt.ylabel("Company")
plt.xlabel("Profit")
plt.title("Company and Profit")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Box plot is a standard way of showing the distribution of data based on five-number summary. which is minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It also shows any potential outliers. 

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2], labels=["box1","box2"])
plt.ylabel("Values")
plt.title("Box Plot of Two Distributions")
plt.show()
```
