Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Plots that can be plotted using Seaborn Library:

1)Scatter Plot: Used to visualize the relationship between two continuous variables by plotting data points on a two-dimentional graph.

2)Dist Plot: Used to visualize a single continuous variable, to give insights into the data's spread, central tendency and shape.

3)Rel Plot: Used for visualization of relationship between multiple variables, supporting both scatter and line plot.

4)Cat Plot:Used to visualize categorical data through various plot types like bar, box, strip plots.

5)Pair Plot:Gives Each variable against each variable and present them in a single graph.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
```

```python
fmri1 = sns.load_dataset("fmri")
```

```python
fmri1.head()
```

```python
fmri1["event"].value_counts()
```

```python
fmri1["region"].value_counts()
```

```python
sns.lineplot(x="timepoint",y="signal", data=fmri1, hue="region",style="event")
```
